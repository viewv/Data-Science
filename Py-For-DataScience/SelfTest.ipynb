```python
import numpy as np
import pandas as pd
import random
```

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
```

```python
frame = pd.DataFrame(data)
```

```python
frame
```

```python
frame.shape
```

```python
df = frame.reindex(['state','pop'],axis=1)
```

```python
df
```

```python
df['random'] = 0
```

```python
f = lambda x: random.randint(1,365) 
```

```python
df['random'] = df['random'].apply(f)
```

```python
ans = df.reindex(['random'],axis=1)
```
