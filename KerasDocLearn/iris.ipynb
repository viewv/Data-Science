```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
col_names = iris['feature_names'] + ['target']
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns=col_names)
```

```python
data.head()
```

```python
np.random.seed = 42
```

```python
data
```

```python
iris = pd.get_dummies(data,columns=['target'])
```

```python
iris.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(iris, test_size=0.2, random_state=42)
```

```python
train_set.size
```

```python
test_set.size
```

```python
train_set.head()
```

```python
train = train_set.values
```

```python
test = train_set.values
```

```python
train.shape
```

```python
train[0]
```

```python
train_x,train_y = train[:,:3],train[:,4:]
test_x,test_y = test[:,:3],test[:,4:]
```

```python
train_y
```

```python
model = Sequential()
```

```python
model.add(Dense(5, activation='relu',input_shape=(3,)))
```

```python
model.add(Dense(5,activation='relu'))
```

```python
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.fit(train_x, train_y,
          batch_size=10,
          epochs=1000,
          verbose=1,
          validation_data=(test_x, test_y))
```

```python
temp = model.predict(test_x[10:11])
```
